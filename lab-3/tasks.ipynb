```python
import time
import matplotlib.pyplot as plt
from binarytree import bst,Node,build
```

```python
bst_tree = bst(height=3)  

print(bst_tree)
```

```python
def bfs(root):
    if not root:
        return []
    queue = [root]
    result = []
    while queue:
        current = queue.pop(0)
        result.append(current.value)
        if current.left:
            queue.append(current.left)
        if current.right:
            queue.append(current.right)
    return result



print('BFS Traversal:', bfs(bst_tree))
```

```python

def dfs(root):
    result = []
    def _dfs(node):
        if node is not None:
            result.append(node.value)
            _dfs(node.left)
            _dfs(node.right)
    _dfs(root)
    return result



print('DFS Traversal:', dfs(bst_tree))
```

```python
def make_bst(nodes):
    values = list(range(nodes))
    root = build(values)
    return root


sizes = [10,100,1000]
bfs_times = []
dfs_times = []

for n in sizes:
    tree = make_bst(n)
    start = time.time()
    bfs(tree)
    bfs_times.append(time.time() - start)
    start = time.time()
    dfs(tree)
    dfs_times.append(time.time() - start)

print(bfs_times, dfs_times)
```

```python

plt.figure(figsize=(8,5))
plt.plot(sizes, dfs_times, marker='s', label='DFS')
plt.plot(sizes, bfs_times, marker='o', label='BFS')

plt.xlabel('Number of Nodes')
plt.ylabel('Execution Time (seconds)')
plt.title('BFS vs DFS Execution Time on Binary Trees')
plt.legend()
plt.grid(True)
plt.show()
```
