```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as s3x
```

```python
df=pd.read_csv("suiciderate.csv")
```

```python
# df.head()
# df.shape
#df.info()

df.isnull().sum()
```

```python
df["HDI for year"]=df["HDI for year"].fillna(df["HDI for year"].mean())
```

```python
duplicate_data=df[df.duplicated(subset=["country"])]
duplicate_data
```

```python
df.drop_duplicates()
```

```python
#df[" gdp_for_year ($) "]= df[" gdp_for_year ($) "].str.replace(',','').astype(int)
#df["gdp_per_capita ($)"]=df[" gdp_for_year ($) "]/df["population"]
df["gdp_per_capita ($)"]
```

```python
df["suicide to gpt ratio"]=df["suicides_no"] / df["gdp_per_capita ($)"]
df["suicide to gpt ratio"]
```

```python
df = df.drop(columns=["country-year"])
df
```

```python
age_suicide =df.groupby('age')["suicides_no"].sum().sort_values()
age_suicide.plot(kind='barh',title='total suicides by age group')
plt.xlabel('no of suicides')
plt.show()
```

```python
country_suicides=df.groupby("country")["suicides_no"].sum().sort_values(ascending=False)
country_suicides.head().plot(kind='bar',title='suicides by countries')
plt.ylabel('suicides')
plt.show()

country_suicides.tail().sort_values().plot(kind='bar',title='suicides by countries')
plt.ylabel('suicides')
plt.show()
```

```python
s3x.scatterplot(data=df,x=" gdp_for_year ($) ",y='suicides_no')
plt.title('GDP vs Suicides')
plt.show()
```

```python
yearly_trend=df.groupby('year')['suicides_no'].sum()
yearly_trend.plot(kind='line',marker='o',title="suicides trend over the year")
plt.show()
```

```python
yearly_trend=df.groupby('sex')['suicides_no'].sum()
yearly_trend.plot(kind='bar',title="suicides trend over the sex",color=['black','blue'])
plt.ylabel("Number of suy9cides ")
plt.show()
```

```python
yearly_trend=df.groupby('generation')['suicides_no'].sum().sort_values()
yearly_trend.plot(kind='barh',title="suicides trend over the generation")
plt.xlabel("Number of suicides")
plt.show()
```
